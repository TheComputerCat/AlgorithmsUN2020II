```
def se(dic,n,last):
  first = lstr(list(range(1, n + 1)))
  act = first
  pn = 0
  while act != last:
    act = dic[act]
    pn = pn + 1
    if(pn > 100):
      break
  return(pn)

  
```

```
import itertools

def maxn(n):
  m = 0;
  li = list(itertools.permutations( list(range(1, n + 1))))
  for i in li:
    dic = BigPn(list(i))
    s = se(dic, n, lstr(i))
    if (s > m):
      m = s
  return m
```

```
import time
for i in range(2, 14):
  start = time.time()
  m = maxn(i)
  end = time.time()
  mes = "P" + str(i)+" = " +str(m) + ", se calculo en" + str(end - start) + '\n'
  print(mes)
  f = open("punto.txt", "a")
  f.write(mes)
  f.close()
```
